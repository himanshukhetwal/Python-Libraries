```python
#PANDA SERIES
# a series is a 1d array which hold any data type. The axis labels are collectively called the index
```

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([15,16,17])
dic = {1:10,2:20,3:30}
```

```python
pd.Series(my_list) #here indexes are called as labels
```

```python
pd.Series(my_list,index=labels) # here we have put labels as indexes which we have created
```

```python
pd.Series(dic)
```

```python
#DATA FRAME
# combination of multiple series is known as data frame
```

```python
#Creation of data frame
data = {
    'Name':['John', 'Anna', 'Peter' ,'Linda'],
    'Location':['New York', 'Paris', 'Berlin', 'London'],
    'Age':[24,13,53,33],
    'Salary': [65000,70000,62000,85000]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
data_list = [
    ['John', 'New York', 24, 65000],
    ['Anna', 'Paris', 13, 70000],
    ['Peter', 'Berlin', 53, 62000],
    ['Linda', 'London', 33, 85000]
]

columns = ['Name', 'Location', 'Age', 'Salary']

df2 = pd.DataFrame(data_list, columns=columns)
```

```python
df2
```

```python
#selection and indexing of columns
df2['Name']
```

```python
df2[['Age','Name']]
```

```python
#Creation of new col
df2['Designation']=['doctor','eng','eng','eng']
df2
```

```python
#remove a col
df2.drop('Age',axis=1)
```

```python
#very imp note axis=1 is used for cols and axis=0 is used for rows
df2
```

```python
#To actually delete col and row from table we use "inplace=True"
df2.drop('Designation', axis=1 , inplace=True)
df2
```

```python
#now droping the rows
df2.drop(0,axis=0)
```

```python
#Selection of rows
df2.loc[0]
```

```python
df2.loc[[0,1]]
```

```python
df2.iloc[0]
```

```python
#Selecting subsets of rows and col
df2.loc[[0,1]][["Location","Salary"]]
```

```python
df2.loc[[2,3]][["Name","Age"]]
```

```python
#Conditional Selection
df2[df2["Age"]>30]
```

```python
df2[(df2['Age']>30) & (df2['Location']=='Berlin')]
```

```python
#HANDLING MISSING DATA
```

```python
data = {
    'A' : [1,2,np.nan,4,5],
    'B' : [1,2,3,4,5],
    'C' : [1,2,3,np.nan,np.nan],
    'D' : [1,np.nan,np.nan,np.nan,5]
} #np.na are null values
df3=pd.DataFrame(data)
df3
```

```python
#Finding Missing Data
df3.isna()
```

```python
df3.isna().sum()#it counted how many null values are in columns
```

```python
df3.isna().any() #means each col has null values
```

```python
#Removing NUll Values or Missing Data
#it works on row basis
df3.dropna() # all the values in row are droped where they have null values except
#the second row
```

```python
df3.dropna(thresh=3) #thresh is like a limit means
#remove rows except which has 3 non null values
```

```python
#Filling The Missing Data
```

```python
#filling null values with 0
df3.fillna(0)
```

```python
#filling missing data with different values in each col
values={'A':0,'B':100,'C':200,'D':300}
df3.fillna(value=values)
```

```python
#filling values with mean
df3.fillna(df3.mean())
```
