```python
#valid SUDOKU
```

```python
import numpy as np
```

```python
s=np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],

    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],

    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
])
```

```python
#suduko is valid when its sum comes 45 for each row and col
#and also valid when its 3*3 matrix sum also should comes as 45
```

```python
b=np.sum(s,axis=1)#rows axis is 1 for sum
b
```

```python
b==45
```

```python
#another way for rows using loop
for i in b:
  if i !=45:
    print("not valid sudoku")
else:
  print("valid Sudoku")
```

```python
#now for doing cloumns
c=np.sum(s,axis=0)#columns axis =1 for sum
c
```

```python
c==45
```

```python
#another way for rows using loop
for i in c:
  if i !=45:
    print("not valid sudoku")
else:
  print("valid Sudoku")
```

```python
#now checking suduko for 3*3 matrix
for i in range(0,9,3):
  for j in range(0,9,3):
    n=s[i:i+3,j:j+3]
    print(n.sum())
```

```python
#Some General Questions of NUMPY ARRAY
```

```python
# columns[AGE,MATH MARKS,SCIENCE MARKS]
data = np.array([
    [18,85,78],
    [19,92,88],
    [17,76,95],
    [18,65,78],
    [20,90,85]
    ])
```

```python
np.mean(data[:,0]) # avg of age
```

```python
data[:,1] #extracting maths marks of all studnets
```

```python
np.max(data[:,2]) #getting highest science marks
```

```python
# get details of student who scored more than 90 in maths
data[data[:,1] > 90]
```

```python
#increase  maths marks of all students by 5
data[:,1] + 5
```

```python
#find how many students are yonger than 19
len(data[data[:,0] < 19])
```

```python
#calculate the avg marks of each subject(col-wise mean)
np.mean(data[:,1:3],axis=0)
```

```python
#get data of students who scored at least 80 in both subjects
data[(data[:,1]>=80) & (data[:,2]>=80)]
```

```python
#replace all science marks < 75 with 0
data[:,2][data[:,2] < 75]=0
```
